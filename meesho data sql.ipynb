```python
df = pd.read_csv(r"C:\Users\mythr\Data&analysis\meesho_data.csv")
print(df.head())
print(df.dtypes)
```

```python
df.fillna('', inplace=True)
```

```python
try:
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="root",  # replace if different
        database="product_catalog"
    )
    cursor = conn.cursor()

    for _, row in df.iterrows():
        sql = """
        INSERT INTO product_data (
            scrape_date, product_name, price, original_price, discount_percentage,
            category, product_url, rating, review_count, seller_name, gender,
            size, state, city, quantity
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        """
        values = (
            row['scrape_date'], row['product_name'], row['price'], row['original_price'],
            row['discount_percentage'], row['category'], row['product_url'], row['rating'],
            row['review_count'], row['seller_name'], row['gender'], row['size'],
            row['state'], row['city'], row['quantity']
        )
        try:
            cursor.execute(sql, values)
        except Exception as e:
            print("❌ Error inserting row:", e)

    conn.commit()
    print("✅ Data imported successfully!")
    cursor.close()
    conn.close()

except mysql.connector.Error as err:
    print("❌ MySQL connection error:", err)
```
